```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

3 Load datset

```python
df=pd.read_csv("diabetes.csv")
df.head(5)
```

4 Exploring the data

```python
print("Shape:", df.shape)
print(df.info())
print(df.isnull().sum())
print(df.describe())
```

).

6 Analysis with Numpys and Pandas

```python
# Average Age of patients 
avg_age=df['Age'].mean()
print("Average Age of Patients:",avg_age)

#Diabetes rate by Pregnancies
diabetes_by_pregnancies = df.groupby('Pregnancies')['Outcome'].mean()
print("Diabetes Rate by Pregnancies:\n", diabetes_by_pregnancies)

#Average BMI of patient with Diabetes
avg_bmi = df[df['Outcome'] == 1]['BMI'].mean()
print('Average BMI of Diabetes Patients:', avg_bmi)

#Average Glucose Level of PAtient with Duabetes
avg_glcs=df[df['Outcome']==1]['Glucose'].mean()
print("Average Glucose Levels of Patients",avg_glcs)
# Diabetes rate by Age
diabetes_by_age = df.groupby('Age')['Outcome'].mean()
print("Diabetes Rate by Age:\n", diabetes_by_age)

```

 Insights:
 
1.Average age: ~33 years.
 
2.Diabetes is increased after age of 29

3 50% chance of getting diabetes for a person who is having more than 6 pregnencies


7 visualization by Matplot

```python
diabetes_rate_by_age = df.groupby('Age')['Outcome'].mean()

plt.figure(figsize=(10,6))
plt.plot(diabetes_rate_by_age.index, diabetes_rate_by_age.values, color='darkorange', marker='o')
plt.title('Diabetes Rate by Age')
plt.xlabel('Age')
plt.ylabel('Diabetes Rate')
plt.grid(True)
plt.show()
```

```python
diabetes_rate_by_preg = df.groupby('Pregnancies')['Outcome'].mean()

plt.figure(figsize=(8,5))
plt.plot(diabetes_rate_by_preg.index, diabetes_rate_by_preg.values, marker='o', color='teal')
plt.title('Diabetes Rate by Number of Pregnancies')
plt.xlabel('Number of Pregnancies')
plt.ylabel('Diabetes Rate')
plt.grid(True)
plt.show()
```

8 visualization with seaborn

```python
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df[['Age', 'SkinThickness', 'Outcome']], hue='Outcome')
plt.show()
```
